```python
def Model_Building(name,model,X_train,X_test,Y_train,Y_test):
    model.fit(X_train,Y_train)
    pred=model.predict(X_test)
    accuracy=accuracy_score(pred,Y_test)
    print("Predictions for {} ".format(name))
    print("Confusion Marix:")
    con_mat=confusion_matrix(pred,Y_test)
    plot=sns.heatmap(con_mat,annot=True,cmap='Blues')
    plot.set_xlabel('Predicted')
    plot.set_ylabel('Actual')
    plt.show()
    print("Accuracy:   {}".format(accuracy))
    precision=precision_score(pred,Y_test)
    recall =recall_score(pred,Y_test)
    F1_score=2*precision*recall/(precision+recall)
    print("precision: {}".format(precision))
    print("Recall: {}".format(recall))
    print("F1_score: {}".format(F1_score))
    print('\n')
    
    
for name,model in models.items():
    Model_Building(name,model,X_train,X_test,y_train,y_test)
```
